#### Write a python function to ﬁnd µ and σ: 

$N(x|{\mu,\sigma}) = N(x|{\mu_1,\sigma_1})N(x|\mu_2,\sigma_2)...N(x|\mu_n,\sigma_n)$   

#### In which $N$ is a Normal distribution:

$N(x|{\mu,\sigma})= \frac{1}\{2{\pi}{\sigma}^2$




#### Your function should take two vectors:


${\mu}= [{\mu_1,\mu_2...\mu_n}]  
{\sigma}= [{\sigma_1,\sigma_2...\sigma_n}]$ 

####            And return µ and σ.






The product of two gaussians is another gaussian (unnormalized) with ${\mu}$ and ${\sigma}$





This can be generalised to a product of N gaussians with distributions ${X_n}$ ∼ ${N}$(${\mu_n}$,${\sigma_n}$), where 







#### In order to calculate the values of ${\mu}$ and ${\sigma}$ , first, we are going to generate some random arrays to represent the vectors 

${\mu}= [{\mu_1,\mu_2...\mu_n}] ,     
        {\sigma}= [{\sigma_1,\sigma_2...\sigma_n}]$ 

```python
import numpy as np


#use the numpy.random.random_integers()to generate a random n-dimensional array of integers.

## testing values 
#mu = np.array([1,2])
#delta = np.array([3,4])

mu = np.random.randint(1,10, 5)
print("1-dimensional random integer array \n", mu,"\n")
delta = np.random.randint(1,10,5)
print("1-dimensional random integer array \n", delta,"\n")
deltasquare=np.square(delta)
dimension=len(delta)
len(delta)
print("Dimension of our Vectors \n", dimension)

```

####            Once we have our vectors defined, we can build a function that returns µ and σ, given those two input vectors and following the product of N gaussians with distributions 
${X_n}$ ∼ ${N}$(${\mu_n}$,${\sigma_n}$)



```python

def d(x, y):

    numerator_mu=np.sum(x / (2 * y))
    denominator_mu=np.sum(1 / (2 * y))
    mu_result=numerator_mu/denominator_mu
    delta_result = np.sqrt(np.prod(y) / np.sum(y))

    return mu_result,delta_result

dResult=d(mu,deltasquare)

print("The µ and σ as a result of a product of ", dimension, " Normal distributions with vector  µ=[µ1,µ2,...,µN]=", mu,
      "and vector σ=[σ1,σ2,...,σN]=",deltasquare, "is", dResult)
```

#### <span style="color:#000066">**Start by sampling potential values for the vectors_X**<span>

```python
import numpy as np
import matplotlib.pyplot as plt

mu = np.random.randint(1,10, 5)
delta = np.random.randint(1,10,5)

def Gaussian_prob(x, _mu=0, _sigma=1):
    """Gaussian distribution: probability of getting x given mu=_mu, sigma=_sigma"""
    return (1/np.sqrt(2*np.pi*_sigma**2))*np.exp(-(x-_mu)**2/(2*_sigma**2))
```

```python
x_span = np.linspace(-10, 10, 1000)
for i in range(1,5):
    prob_list = [Gaussian_prob(x, mu[i], delta[i]) for x in x_span]
    print(prob_list)
```

```python

plt.figure()
plot_list1 = plt.plot(x_span, prob_list)
for i in plot_list1:
    plt.plot()
```
